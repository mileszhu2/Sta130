Work with a ChatBot to create code to produce a 95% bootstrap confidence interval for a population mean based on a sample that you have and comment the code to demonstrate how the code can be changed to produce a 95% bootstrap confidence interval for different population parameter (other than the population mean, such as the population median).<br>

````python
# Code produced by ChatGPT explained by me

import pandas as pd # python library for data manipulation and analysis
import numpy as np # python library that allows for numerical operations on large datasets

# Load the dataset
url = "https://github.com/rfordatascience/tidytuesday/raw/master/data/2024/2024-01-09/canada_births_1991_2022.csv"
data = pd.read_csv(url)

# We will use the 'births' column for this analysis
# This means we calculate for monthly births
# We will drop all rows with missing values is the 'births' column
data_column = data['births'].dropna() 

# Function to calculate bootstrap confidence interval
# Parameters: data(input data), statistic_function(python function), n_iterations(number of bootstrapped samples), alpha(a number to find percentiles) 
def bootstrap_ci(data, statistic_function, n_iterations=1000, alpha=0.05): 
  
    # Generate bootstrap samples with replace=True allowing for sampling with replacement
    # bootstrap_samples is an arrary where each row is a bootstrapped sample of data
    # It will generate 1000 samples (n_iterations), each sample being the same length as the original sample of data
    bootstrap_samples = np.random.choice(data, (n_iterations, len(data)), replace=True)
    # Applies the python function on each bootstrapped sample to find the desired statistic for each sample
    bootstrap_statistics = np.array([statistic_function(sample) for sample in bootstrap_samples])
    
    # Calculate the lower and upper percentiles for the confidence interval
    lower_bound = np.percentile(bootstrap_statistics, 100 * (alpha / 2)) # The 2.5 percentile
    upper_bound = np.percentile(bootstrap_statistics, 100 * (1 - alpha / 2)) # The 97.5 percentile
    
    # Returns a tuple containing the lower and upper bounds of the interval
    return lower_bound, upper_bound

# Calculate the 95% bootstrap confidence interval for the mean using the function we created
# The data is the 'births' column with the useless rows dropped and the function is np.mean which finds the mean of a dataset
# Prints the results
mean_ci = bootstrap_ci(data_column, np.mean)
print(f"95% Bootstrap Confidence Interval for Mean: {mean_ci}")

# Calculate the 95% bootstrap confidence interval for the median using the function we created
# The data is the 'births' column with the useless rows dropped and the function is np.median which finds the median of a dataset
# Prints the results 
median_ci = bootstrap_ci(data_column, np.median)
print(f"95% Bootstrap Confidence Interval for Median: {median_ci}")
````

````python
import numpy as np # Python library for numerical operations on large datasets
import pandas as pd # Python library for data manipulation and analysis
import matplotlib.pyplot as plt # Python library for creation of visuals that represent data

# Load the Dataset
df = pd.read_csv("STA130 Vaccine Data.csv")

# Calculate the change in health scores
# Create a new array that is the difference in the values of the 'FinalHealthScore' and 'InitialHealthScore' columns for each row
df['Change'] = df['FinalHealthScore'] - df['InitialHealthScore']

# Bootstrap parameters
n_iterations = 10000 # Amount of bootstrap samples
confidence_level = 0.95 # The confidence percent for the confidence interval (95% in this case)
n_samples = len(df) # Amount of data in each sample is the number of rows in the original dataset

# Bootstrap sampling
boot_means = [] # Create an empty array for the means of each bootstrap sample
np.random.seed(42)  # For reproducibility

for _ in range(n_iterations): # For each index in n_iterations (0-9999 or in other words the following code will repeat 10000 times)
    # Creates a bootstrap sample picking random values from our array of the change in health scores
    # The amount of rows in this sample is the same as our original dataset (10) and replace=True allows for sampling with replacement
    sample = np.random.choice(df['Change'], size=n_samples, replace=True)  
    # The bootstrap sample mean is added to the array boot_means
    boot_means.append(np.mean(sample))

# Calculate confidence intervals
lower_bound = np.percentile(boot_means, (1 - confidence_level) / 2 * 100) # Finds the 2.5 percentile (lower bound of confidence interval)
upper_bound = np.percentile(boot_means, (1 + confidence_level) / 2 * 100) # Finds the 97.5 percentile (Upper bound of confidence interval)

# Print confidence interval
print(f"{confidence_level*100}% Confidence Interval for the Mean Change in Health Score: ({lower_bound:.2f}, {upper_bound:.2f})")

# Plotting
plt.figure(figsize=(10, 6)) # Size of the figure/graph
plt.hist(boot_means, bins=30, alpha=0.7, color='skyblue', edgecolor='black') # Create a 30 bin histogram out of the values in boot_means
plt.axvline(lower_bound, color='red', linestyle='dashed', linewidth=2, label='Lower CI Bound') # Line for lower bound of 95% confidence interval
plt.axvline(upper_bound, color='green', linestyle='dashed', linewidth=2, label='Upper CI Bound') # Line for upper bound of 95% confidence interval
plt.title('Bootstrap Distribution of Mean Change in Health Score') # Title of graph
plt.xlabel('Mean Change in Health Score') # X-axis label
plt.ylabel('Frequency') # Y-axis label
plt.legend() # Adds a legend
plt.grid() # Adds gridlines
plt.show() # Displays the graph
````

Findings and Discussion:
- From the histogram of the means of the bootstrapped samples, we can observe that the confidence interval is positive and does not include zero. This is similar to the graph on the right that I showed earlier. Because the confidence interval does not include zero, I can reject the null hypothesis and say that the vaccine most likely has a positive effect.
- I can reject the null hypothesis because the confidence interval tells me that I can be 95% sure that the real average effect lies in that interval. Since all possibilities in that interval are positive, I can be 95% sure that the vaccine has a positive effect.
